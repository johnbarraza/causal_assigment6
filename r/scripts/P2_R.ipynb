```r
#Agregación event-time
df$event_time <- df$t - df$g

agg_event <- aggregate(asmrs ~ event_time, data = df, FUN = mean)
agg_event
```

**Agregación por grupo (cohort):**  
Agrupa las observaciones por año de tratamiento \(g\) y calcula el promedio del resultado dentro de cada cohorte. Esto permite comparar si los estados que se tratan más temprano muestran niveles de asmrs distintos de los que se tratan más tarde.

**Agregación por período:**  
Agrupa las observaciones por año calendario \(t\), promediando asmrs en cada período. De esta forma se observa la evolución temporal del resultado en el tiempo, sin distinguir aún por cohorte.

**Agregación por event-time:**  
Utiliza la variable \(event\_time = t - g\), donde los valores negativos representan años previos al tratamiento, el valor 0 es el año en que empieza el tratamiento y los valores positivos son años posteriores. Al promediar asmrs por event-time, se alinean todas las unidades en torno al momento del tratamiento, lo que permite estudiar la dinámica antes y después del mismo.

**¿Qué agregación es más comparable al Event-Study del modelo TWFE?**  
La agregación por event-time es la más comparable al Event-Study TWFE, porque en ambos casos los efectos se organizan en función de los períodos relativos al inicio del tratamiento (leads y lags), en lugar de hacerlo por año calendario o por cohorte directamente.

```r
df2 <- df   # versión simple sin recorte

twfe_model <- lm(
  asmrs ~ factor(event_time) + factor(id) + factor(t) +
    pcinc + asmrh + cases,
  data = df2
)

summary(twfe_model)
```

```r
coefs       <- coef(twfe_model)
names_coefs <- names(coefs)
```

```r
idx_event <- grepl("^factor\\(event_time\\)", names_coefs)

twfe_event <- data.frame(
  variable = names_coefs[idx_event],
  coef     = coefs[idx_event]
)
```

```r
twfe_event$event_time <- as.numeric(
  sub("factor\\(event_time\\)", "", twfe_event$variable)
)
```

```r
twfe_event <- twfe_event[order(twfe_event$event_time), ]

head(twfe_event)
```
